# 6. Preparación de los Datos

```python
data_clean = data.drop(columns = ['Customer ID', 'Senior Citizen', 'Country', 'State', 'Quarter', 'Latitude', 'Longitude', 'Total Charges', 'Total Revenue', 'Offer', 'Churn Category', 'Churn Reason', 'CLTV', 'Customer Status', 'City', 'Satisfaction Score', 'Churn Score'])
data_clean
```

## Imputación

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import make_column_transformer
```

```python
categorical_cols = data_clean.select_dtypes(include = ['object']).columns
categorical_transformer = make_column_transformer((OrdinalEncoder(), categorical_cols),remainder='passthrough')
categorical_transformer
```

```python
data_transformed = pd.DataFrame(categorical_transformer.fit_transform(data_clean),
                                columns = categorical_transformer.get_feature_names_out(),
                                index = data_clean.index).rename(columns = lambda x: x.removeprefix('ordinalencoder__')).rename(columns = lambda x: x.removeprefix ('remainder__'))

data_transformed
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
recursive_imputer = IterativeImputer(estimator= LinearRegression(),
                                     initial_strategy= 'mean',
                                     max_iter=10)

data_recursive = pd.DataFrame(recursive_imputer.fit_transform(data_transformed).round(),
                              columns= data_transformed.columns,
                              index= data_transformed.index)

data_recursive.info()
```

```python
enc  = categorical_transformer.named_transformers_.ordinalencoder
cats = enc.categories_

for i, col in enumerate(categorical_cols):
    k = len(cats[i])
    data_recursive[col] = (
        np.rint(data_recursive[col])
        .clip(0, k - 1)               
        .astype("int64")              
    )
```

```python
data_imputed = data_recursive.copy()
data_imputed[categorical_cols] = categorical_transformer.named_transformers_.ordinalencoder.inverse_transform(
    X=data_recursive[categorical_cols]
)
data_imputed.info()
```

## Transformación

```python
data_imputed.to_csv('telco_clean.csv', index=False)
```

```python
X = data_imputed.drop(columns = ['Churn Label'])
y = data_imputed['Churn Label']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
num_col = X.select_dtypes(include = ['int64', 'float64']).columns.to_list()
cat_col = X.select_dtypes(include = ['object']).columns.to_list()

numtrans = Pipeline(steps=[("scale", MinMaxScaler())])
cattrans = Pipeline(steps=[("encoder", OneHotEncoder(sparse_output=False))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numtrans, num_col),
        ('cat', cattrans, cat_col)
    ],
    remainder='passthrough'
)

preprocessor
```

```python
X_pre = preprocessor.fit_transform(X)
cod_cat = preprocessor.named_transformers_['cat']["encoder"].get_feature_names_out(cat_col)
labels = np.concatenate((num_col, cod_cat))
X_pro = pd.DataFrame(X_pre, columns=labels)
indexs = X_pro.columns
X_pro
```

## Conjunto de Entrenamiento

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pro, y, train_size=0.7, stratify=y, random_state=42)
```

## Selección de Rasgos con ShAP

```python
import shap
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_depth=3, random_state=42)
gb.fit(X_train, y_train)
```

```python
explainer = shap.Explainer(gb, X_train, feature_names=indexs)
shap_values = explainer(X_train)
```

```python
shap.plots.bar(shap_values)
```

```python
imp = np.abs(shap_values.values).mean(axis=0)
feature_importance = pd.Series(imp, index=indexs).sort_values(ascending=False)
feature_importance
```

### Selección Modelo

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
```

```python
modelo = GradientBoostingClassifier(random_state=123)

param_grid = {'loss': ['log_loss', 'deviance'], 'n_estimators': [100, 200, 300]}
clf = GridSearchCV(modelo, param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=1)
clf.fit(X_train, y_train)

validation = pd.concat([pd.DataFrame(clf.cv_results_['params']), pd.DataFrame(clf.cv_results_['mean_test_score'], columns=['Accuracy'])], axis=1)
validation
```

```python
modelo = clf.best_estimator_
modelo
```

### Selección Rasgos

```python
import sklearn.metrics as metrics
```

```python
ks = range(1, X_train.shape[1] + 1)
results = []
for k in ks:
    topk = feature_importance.index[:k]
    gb_k = modelo
    gb_k.fit(X_train[topk], y_train)
    pred = gb_k.predict(X_test[topk])
    results.append((k, metrics.accuracy_score(y_test, pred), metrics.f1_score(y_test, pred, average="binary", pos_label="Yes")))

best_k_acc = max(results, key=lambda t: t[1])[0]
best_k_f1  = max(results, key=lambda t: t[2])[0]
```

```python
topk = feature_importance.index[:best_k_f1]
gb_final = modelo
gb_final.fit(X_train[topk], y_train)
pred = gb_final.predict(X_test[topk])

print("Accuracy:", metrics.accuracy_score(y_test, pred))
print("F1:", metrics.f1_score(y_test, pred, average="binary", pos_label="Yes"))
print("Rasgos seleccionados:", list(topk))
```

## 7. Modelos

### Evaluación Modelos

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
modelos = []
```

```python
def evaluar_modelos(models, X_train, Y_train, X_test, Y_test):
    
    for nombre, modelo in models:
        print(f'\n===== {nombre} =====')
        
        # Predicciones
        y_pred_train = modelo.predict(X_train)
        y_pred_test = modelo.predict(X_test)
        
        print('--- Train ---')
        print(classification_report(Y_train, y_pred_train))
        print('Confusion Matrix (Train):')
        print(confusion_matrix(Y_train, y_pred_train))
        
        print('--- Test ---')
        print(classification_report(Y_test, y_pred_test))
        print('Confusion Matrix (Test):')
        print(confusion_matrix(Y_test, y_pred_test))
        
        # Crear subplots para mostrar ambas matrices de confusión
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        # Matriz de confusión Train
        sns.heatmap(confusion_matrix(Y_train, y_pred_train), annot=True, fmt='d', ax=axes[0])
        axes[0].set_title(f'{nombre} - Train')
        axes[0].set_xlabel('Predicted')
        axes[0].set_ylabel('Actual')
        
        # Matriz de confusión Test
        sns.heatmap(confusion_matrix(Y_test, y_pred_test), annot=True, fmt='d', ax=axes[1])
        axes[1].set_title(f'{nombre} - Test')
        axes[1].set_xlabel('Predicted')
        axes[1].set_ylabel('Actual')

        plt.figure(figsize=(6, 3))
        plt.tight_layout()
        plt.show()
```
